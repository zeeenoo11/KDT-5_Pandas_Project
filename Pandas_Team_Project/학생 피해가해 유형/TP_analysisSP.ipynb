```python
# Preparance
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import matplotlib.font_manager as fm # 폰트 관련

# 한글 font 설정
font_path =r'C:\Windows\Fonts\gulim.ttc'
#matplotlib 패키지에 한글폰트 설정
myfont = fm.FontProperties(fname= font_path).get_name()
mpl.rc('font', family = myfont)
```

```python
# 1.   학생 S
# 1.1. 가해 P

fileSV = '사이버폭력_가해_경험_20240119171459.csv'
DFSV = pd.read_csv(fileSV, header=[1], index_col=[0,1])
DFSV
```

```python
# 전체 응답자 중 피해 경험률은 37.5%, 이 중 언어폭력이 가장 높고 명예 훼손, 스토킹 순으로 나타납니다. 


# 전체 비율 비교
dfL0 = DFSV.xs(DFSV.index[0][0], level=0)  # DF 추출 : '전체'


# ------------- 전체 크기 -----------------
plt.figure(figsize=(5,5))


# -------------- 그래프 1 -----------------
WITDH = 5  # 막대 너비

xlabelList = [0,10,20,30,40,50,60,70,80]
x2labelList = list(map(lambda x:x+WITDH,xlabelList))

# ----------- 바 그래프 함수 --------------
plt.bar(
    xlabelList,dfL0.iloc(0)[0], width=WITDH, 
    color=['#F6546A','#045FB4', '#6497b1', '#b3cde0', '#81BEF7', '#A9D0F5', '#CEE3F6', '#81BEF7', '#CEE3F6', "#b3cde0"]
    )

plt.xticks(ticks=xlabelList, labels=dfL0.columns, rotation=315)

# ------------ 제목 설정 ------------------
plt.title("[ 사이버 폭력 가해 경험 ]")

# ------------ 축 다듬기 ------------------
plt.ylabel('%', loc='top', fontsize='large')

# ------------ 범례 설정 ------------------
# plt.legend(['남한', '북한'])

# ------------ 눈금 설정 ------------------
# maxValue = max([max(waterDF[i]) for i in waterDF.columns])

# plt.xticks(ticks=[i for i in range(0, 30, 5)])
# plt.yticks(ticks=[i for i in range(0, maxValue, maxValue//10)])

# ------------ 격자 설정 ------------------
# plt.grid(True, lw =0.5, ls ='-.', c='.9')
```

```python
# 성별을 기준으로 보면 남학생의 피해 경험률은 42.1%, 여학생의 피해 경험률은 32.5% 이며 항목별 순서는 전체 비율과 동일합니다.


# 전체 비율 비교
dfL1 = DFSV.xs(DFSV.index[1][0], level=0)  # DF 추출 : '성별'

# ------------- 전체 크기 -----------------
plt.figure(figsize=(10,5))


# -------------- 그래프 1 -----------------
WITDH = 5  # 막대 너비

xlabelList = [0,15,30,45,60,75,90,105,120]
x2labelList = list(map(lambda x:x+WITDH,xlabelList))

# ----------- 바 그래프 함수 --------------
plt.bar(
    xlabelList,dfL1.iloc(0)[0], width=WITDH, 
    color=['#045FB4', '#6497b1', '#b3cde0', '#81BEF7', '#A9D0F5', '#CEE3F6', '#81BEF7', '#CEE3F6', "#b3cde0"]
    )
plt.bar(
    x2labelList,dfL1.iloc(0)[1], width=WITDH, 
    color=[ '#D32F2F','#E57373','#FFCDD2', '#EF9A9A', '#FF8A80','#EF9A9A', '#FFCDD2','#FF8A80', '#EF9A9A']
    )

# ------------ 눈금 설정 ------------------
plt.xticks(ticks=xlabelList, labels=dfL1.columns, rotation=315)

# ------------ 제목 설정 ------------------
plt.title("[ 사이버 폭력 가해 경험 ]")

# ------------ 축 다듬기 ------------------
plt.ylabel('%', loc='top', fontsize='large')

# ------------ 범례 설정 ------------------
plt.legend(['남학생', '여학생'])

# ------------ 눈금 설정 ------------------
# maxValue = max([max(waterDF[i]) for i in waterDF.columns])

# plt.xticks(ticks=[i for i in range(0, 30, 5)])
# plt.yticks(ticks=[i for i in range(0, maxValue, maxValue//10)])

# ------------ 격자 설정 ------------------
# plt.grid(True, lw =0.5, ls ='-.', c='.9')
```
